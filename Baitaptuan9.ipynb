```python
#W8A1
class Cylinder:
    def __init__(self,bankinh,chieucao):
        self.bankinh = bankinh
        self.chieucao = chieucao    
    def thetich(self):
        pi = 3.14
        return pi * (self.bankinh ** 2) * self.chieucao
    def dientich(self):
        pi = 3.14
        return 2 * pi * self.bankinh * (self.bankinh + self.chieucao)
bankinh1 = float(input("Nhập bán kính hình trụ: "))
chieucao1 = float(input("Nhập chiều cao hình trụ: "))
Cylinder1 = Cylinder(bankinh1, chieucao1)
print(f"Thể tích là {Cylinder1.thetich():.2f}")
print(f"Diện tích là {Cylinder1.dientich():.2f}")        
```

```python
#W8A2
class Calender:
    def __init__(self,date,month,year):
        self.date = date
        self.month = month
        self.year = year
    def kiemtra(self):
        ngay = [31,28,31,30,31,30,31,31,30,31,30,31]
        if (self.year % 4 == 0 and self.year % 100 != 0) or (self.year % 400 == 0):
            ngay[1] = 29
        if 1 <= self.month <= 12:
            if 1 <= self.date <= ngay[self.month - 1]:
                return True
            else:
                return False
        else:
            return False
    def ngayke(self):
        ngay = [31,28,31,30,31,30,31,31,30,31,30,31]
        if (self.year % 4 == 0 and self.year % 100 != 0) or (self.year % 400 == 0):
            ngay[1] = 29
        self.date += 1
        if self.date > ngay[self.month - 1]:
            self.date = 1
            self.month += 1
            if self.month > 12:
                self.month = 1
                self.year += 1
        return (self.date, self.month, self.year)
d,m,y = map(int, input("Nhập ngày, tháng, năm (cách nhau bởi dấu cách): ").split())
calender1 = Calender(d,m,y)
if calender1.kiemtra():
    print("Ngày hợp lệ")
    ngayke = calender1.ngayke()
    print(f"Ngày kế tiếp là: {ngayke[0]}/{ngayke[1]}/{ngayke[2]}")
else:
    print("Ngày không hợp lệ")
```

```python
#W8A3
import math
class point:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def kiemtra(self):
        if self.x == 0 and self.y == 0:
            return "Gốc tọa độ"
        elif self.x == 0:
            return "Trên trục tung"
        elif self.y == 0:
            return "Trên trục hoành"
        else:
            return "Ngoài trục tọa độ"
    def khoangcach(self):
        return math.sqrt(self.x**2 + self.y**2)
x1,y1 = map(float,input("Nhập tọa độ x,y của điểm ").split())
point1 = point(x1,y1)
print(point1.kiemtra())
print(f"Khoảng cách từ điểm đến gốc tọa độ là: {point1.khoangcach():.2f}")
        
```

```python
#W8A4
class Calculator:
    def __init__(self,a,b):
        self.a = a
        self.b = b
    def add(self):
        return self.a + self.b
    def subtract(self):
        return self.a - self.b
    def product(self):
        return self.a * self.b
    def divide(self):
        if self.b != 0:
            return self.a / self.b
        else:
            return "Không thể chia cho 0"
    def power(self):
        return self.a ** self.b
    def mod(self):
        return self.a % self.b
    def set_number(self,a,b):
        self.a = a
        self.b = b
while True:
    tieptuc = input("Bạn có muốn tiếp tục không?")
    if tieptuc.lower() not in ["yes", "YES", "Yes"]:
        break
    a,b = map(float, input("Nhập hai số a và b (cách nhau bởi dấu cách): ").split())
    calc = Calculator(a,b)
    print("Menu:")
    print("1. Cộng")
    print("2. Trừ")
    print("3. Nhân")
    print("4. Chia")
    print("5. Lũy thừa")
    print("6. Chia lấy dư")
    print("7. Thay đổi số")
    choice = int(input("Nhập tùy chọn 1-7"))
    if choice == 1:
        print("Kết quả:", calc.add())
    elif choice == 2:
        print("Kết quả:", calc.subtract())
    elif choice == 3:
        print("Kết quả:", calc.product())
    elif choice == 4:
        print("Kết quả:", calc.divide())
    elif choice == 5:
        print("Kết quả:", calc.power())
    elif choice == 6:
        print("Kết quả:", calc.mod())
    elif choice == 7:
        a,b = map(float, input("Nhập hai số a và b (cách nhau bởi dấu cách): ").split())
        calc.set_number(a,b)

    
```
