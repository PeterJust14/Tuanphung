```python
#Cau13
a,b = map(int,input().split())
tonguoca = 0
tonguocb = 0
for i in range(1,a):
    if a % i==0:
        tonguoca += i
for i in range(1,b):
    if b % i ==0:
        tonguocb += i
if tonguoca == b and tonguocb == a:
    print("dung")
else:
    print("sai")
```
