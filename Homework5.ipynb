```python
#W6A20
#Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và 
#trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về 
#-1
def find(ds, k):
    for i in range(len(ds)):
        if ds[i] == k:
            return i + 1
            
    return -1
ds1 = list(map(int, input("Nhập dãy số nguyên cách nhau dấu cách:").split()))
k1 = int(input("Nhập số nguyên dương k:"))
print(find(ds1, k1))
```

```python
#W6A23 VIết hàm xóa đi các số <x có trong mảng
def xoa_so_nho_hon_x(ds, x):
    kq = []
    for i in ds:
        if i >= x:
            kq.append(i)
    return kq
ds1 = list(map(int, input("Nhập dãy số nguyên cách nhau dấu cách:").split()))
x1 = int(input("Nhập x:"))
print(xoa_so_nho_hon_x(ds1,x1))
```

```python
#W6A24
#Viết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí 
#trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9
def day_so_0_ve_dau(ds):
    ds1 = []
    ds2 = []
    for so in ds:
        if so == 0:
            ds1.append(so)
        else:
            ds2.append(so)
    dscuoi = ds1 + ds2
    for a in dscuoi:
        print(a, end=' ')
danhsach = list(map(int, input("Nhập dãy số nguyên cách nhau dấu cách:").split()))
day_so_0_ve_dau(danhsach)
```

```python
#W6A26
#Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2 
#phần tử đầu và cuối
ds = list(map(int, input("Nhập dãy số nguyên cách nhau dấu cách:").split()))
tong = 0
for i in range(1, len(ds)-1):
    if ds[i] > ds[i-1] and ds[i] > ds[i+1]:
        tong += ds[i]
print(f"Tổng các phần tử cực đại địa phương là: {tong}")
```
