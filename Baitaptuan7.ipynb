```python
'''WViết hàm tìm kiếm nhị phân
trên một danh sách đã
được sắp xếp tăng dần.
Trả về chỉ số của phần tử
hoặc -1 nếu không tìm
thấy. mẫu hàm: def
binary_search(arr, left,
right, target):'''
def binary_search(arr, left, right, target):
    while left <= right:
        mid = left + (right - left) // 2

        # Kiểm tra nếu target có ở giữa
        if arr[mid] == target:
            return mid
        # Nếu target lớn hơn, bỏ qua nửa bên trái
        elif arr[mid] < target:
            left = mid + 1
        # Nếu target nhỏ hơn, bỏ qua nửa bên phải
        else:
            right = mid - 1

    # Target không có trong mảng
    return -1
```

```python
'''W7A2
Cho danh sách arr và một
số nguyên x, đếm xem x
xuất hiện bao nhiêu lần
trong danh sách. def
count_occurrences(arr, x) -
> int:'''
def count_occurrences(arr, x):
    count = 0
    for element in arr:
        if element == x:
            count += 1
    return count
```

```python
'''W7A3
Viết hàm thực hiện thuật
toán bubble sort, Selection
Sort để sắp xếp danh sách
theo thứ tự tăng dần. Sắp
xếp nổi bọt (Bubble Sort)'''
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
def selection_sort(arr):
    A=[]
    for i in range(len(arr)):
        somin = min(arr)
        A.append(somin)
        arr.remove(somin)
    return A
print(bubble_sort([64, 34, 25, 12, 22, 11, 90]))
print(selection_sort([64, 34, 25, 12, 22, 11, 9]))


    
        
```

```python
#W7A5
'''Cho danh sách các số
nguyên và một số X, đếm
tất cả các cặp (a, b) sao
cho a + b = X. c
'''
def giaitoan(dayso, X):
    A=set()
    n = len(dayso)
    for i in range(n):
        for j in range(i+1,n):
            if (dayso[i] + dayso[j]) % X ==0:
                cap = (dayso[i], dayso[j])
                A.add(cap)
                
                
    return A
day = [1,2,3,4,5,6,7,8,9]
print(giaitoan(day, 10))
```

```python
#W7A13
day = list(map(int, input("Nhập các số nguyên, cách nhau bằng dấu cách: ").split()))
n = len(day)
A = []
for i in range(n):
    for j in range(i+1, n+1):
        mangcon = day[i:j]
        A.append(mangcon)

B = []
for mang in A:
    sorted_mang = sorted(mang)
    if sorted_mang[-1] - sorted_mang[0] == 1:
        B.append(mang)

max_len = 0
for mangcuoi in B:
    if len(mangcuoi) > max_len:
        max_len = len(mangcuoi)

print(max_len)
```
