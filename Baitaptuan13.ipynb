```python
"""Bất cứ một số nguyên dương nào cũng tồn tại một số nguyên dương là số đảo ngược của nó. 

Hãy viết hàm def reverse (path: str) -> int: để tìm số nguyên đảo ngược của số n 

Đầu vào 
Đầu vào của hàm là một xâu biểu diễn đường dẫn đến tệp. Nội dung tệp chứa duy nhất một số nguyên dương. 

Đầu ra 
Hàm trả về số nguyên dương đảo ngược của số trong tệp đầu vào."""
def reverse(path: str) -> int:
    with open(path, 'r') as file:
        number_str = file.read().strip()
    reversed_str = number_str[::-1]
    return int(reversed_str)
print(reverse('c1.txt'))
```

```python
'''Viết hàm def count(path: str, character: str), nhận tham số truyền vào là một tên file và một kí tự character, trả về số lần xuất hiện của kí tự character trong chuỗi sentence đọc được từ file.
'''
def count(path, cha):
    with open(path, 'r') as file:
        content = file.read()
    return content.count(cha)
```

```python
'''Viết chương trình in các từ Fibonacci thứ 0 đến n (n<=20) và ghi ra file output.txt. Công thức của dãy Fibonacci như sau:

f(0)=a
f(1)=b
f(n)=f(n−1)+f(n−2)'''
def fibonacci_to_file(a, b, n):
    fib_sequence = [a, b]
    for i in range(2, n):
        next_fib = fib_sequence[i-1] + fib_sequence[i-2]
        fib_sequence.append(next_fib)
    
    with open('output.txt', 'w') as file:
        for num in fib_sequence[:n]:
            file.write(f"{num}\n")
```

```python
'''Bạn là một kẻ trộm chuyên nghiệp và đang lên kế hoạch trộm những ngôi nhà dọc theo một con phố. 

Mỗi ngôi nhà có một số tiền nhất định, sự ràng buộc duy nhất ngăn cản bạn lấy trộm là hai nhà liền kề có hệ thống an ninh kết nối và nó sẽ tự động liên lạc với cảnh sát nếu hai nhà liền kề đã bị đột nhập vào cùng một đêm. 

Cho danh sách các số nguyên dương biểu diễn cho số tiền của mỗi ngôi nhà, xác định số tiền tối đa bạn có thể trộm tối nay mà không bị báo cảnh sát. 

Viết hàm def getMoney (path: str) -> list[int] để đọc số tiền chứa trong mỗi ngôi nhà từ tệp đầu vào được lưu theo đường dẫn path và trả về mảng chứa các số tiền này 
. Số tiền của mỗi nhà được lưu trong file cách nhau bởi một khoảng trắng và không biết trước số lượng nhà trong khu phố đó. Hàm trả về mảng chứa số tiền của các nhà trong khu phố. 

Viết hàm def rob(hourses: list[int]) -> int để tính và trả về số tiền lớn nhất mà bạn có thể trộm được từ con phố đó mà không bị phát hiện'''
def getMoney (path: str) -> list[int]:
    with open(path, 'r') as file:
        content = file.read().strip()
    return list(map(int, content.split()))
def rob(hourses: list[int]) -> int:
    n = len(hourses)
    if n == 0:
        return 0
    elif n == 1:
        return hourses[0]
    
    dp = [0] * n
    dp[0] = hourses[0]
    dp[1] = max(hourses[0], hourses[1])
    
    for i in range(2, n):
        dp[i] = max(dp[i-1], dp[i-2] + hourses[i])
    
    return dp[-1]
```

```python
'''Với một dãy số nguyên cho trước, hãy viết hàm def maximumProduct (path: str) -> int: tìm bộ ba số có tích lớn nhất trong tất cả các bộ số và trả về tích lớn nhất đó.

Hàm nhận đầu vào là đường dẫn đến tệp lưu mảng số nguyên cho trước. Hàm trả về tích lớn nhất có thể có.

Dòng đầu tiên của tệp chứa kích cỡ n, dòng kế tiếp chứa n số nguyên cách nhau bởi một khoảng trắng.'''
def maximumProduct (path: str) -> int:
    with open(path, 'r') as file:
        lines = file.readlines()
        n = int(lines[0].strip())
        numbers = list(map(int, lines[1].strip().split()))
        
    numbers.sort()
    return max(numbers[-1] * numbers[-2] * numbers[-3], numbers[0] * numbers[1] * numbers[-1])
```

```python
'''Thành phố X mở một rạp chiếu phim mới, rất nhiều người muốn đến xem thử và hưởng các khuyến mại trong ngày khai trương.

Để đưa cho khách hàng các lựa chọn, rạp chiếu phim treo các poster về phim cùng với điểm đánh giá IMDB (rating) và mô tả của khách hàng về bộ phim đó.

Hãy viết hàm def findMovies (path: str) -> list[str] giúp khán giả tìm ra phim với điểm đánh giá lớn hơn 8.0 và có mô tả Không phải là "nhàm chán" ("boring").

Hàm nhận đầu vào là đường dẫn đến tệp chứa các poster phim. Hàm trả lại các tên những phim thỏa mãn điều kiện trên. Nếu không tìm thấy phim nào phù hợp, trả về thông báo "All are boring!". Kết quả in theo thứ tự xuất hiện trước sau trong file đầu vào

Tệp có cấu trúc như sau:

id, movie, description, rating
1, Thor, great 3D, 8.0
2, The Rings, Fantasy, 7.5
3, Irish, boring, 6.2
4, Ice song, Fantasy, 8.6
5, Home Alone, Interesting, 6.5'''
def findMovies(path: str) -> list[str]:
    result = []

    with open(path, "r", encoding="utf-8") as f:
        lines = f.readlines()

    # Bỏ dòng tiêu đề
    for line in lines[1:]:
        parts = line.strip().split(",")

        if len(parts) < 4:
            continue

        movie = parts[1].strip()
        description = parts[2].strip().lower()
        rating = float(parts[3].strip())

        if rating > 8.0 and description != "boring":
            result.append(movie)

    if not result:
        return ["All are boring!"]

    return result

```

```python
import math

def print_perimeters(path: str):
    with open(path, "r", encoding="utf-8") as f:
        lines = [line.strip() for line in f if line.strip() != ""]

    i = 0
    while i < len(lines):
        shape_type = lines[i].split(":")[1].strip()
        params = lines[i + 1]

        if shape_type == "SQUARE":
            side = float(params.split(":")[1])
            perimeter = 4 * side

        elif shape_type == "RECTANGLE":
            parts = params.split()
            width = float(parts[1])
            height = float(parts[3])
            perimeter = 2 * (width + height)

        elif shape_type == "CIRCLE":
            radius = float(params.split(":")[1])
            perimeter = 2 * math.pi * radius

        print(f"{perimeter:.2f}")
        i += 2
```

```python
def productExceptSelf(path: str) -> list[int]:
    # Đọc dãy số từ file
    with open(path, "r", encoding="utf-8") as f:
        nums = list(map(int, f.read().split()))

    n = len(nums)

    left = [1] * n
    right = [1] * n
    result = [0] * n

    # Tính tích bên trái
    for i in range(1, n):
        left[i] = left[i - 1] * nums[i - 1]

    # Tính tích bên phải
    for i in range(n - 2, -1, -1):
        right[i] = right[i + 1] * nums[i + 1]

    # Tính kết quả
    for i in range(n):
        result[i] = left[i] * right[i]

    # In ra màn hình
    print(*result)

    return result
```

```python
import math

def estimatedTime(path: str):
    with open(path, "r", encoding="utf-8") as f:
        lines = f.readlines()

    # Dòng SRC
    src_parts = lines[0].split()
    x1 = float(src_parts[2])
    y1 = float(src_parts[4])

    # Dòng DES
    des_parts = lines[1].split()
    x2 = float(des_parts[2])
    y2 = float(des_parts[4])

    # Dòng VELOCITY
    velocity = float(lines[2].split()[1])

    # Khoảng cách Euclid
    distance = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)

    # Thời gian
    time = distance / velocity

    return round(time, 2)
```

```python
def imageSmoother(path: str):
    with open(path, "r", encoding="utf-8") as f:
        lines = f.read().strip().splitlines()

    M = int(lines[0])
    img = [list(map(int, lines[i + 1].split())) for i in range(M)]

    result = [[0] * M for _ in range(M)]

    for i in range(M):
        for j in range(M):
            total = 0
            count = 0

            # duyệt 8 pixel xung quanh và chính nó
            for di in (-1, 0, 1):
                for dj in (-1, 0, 1):
                    ni, nj = i + di, j + dj
                    if 0 <= ni < M and 0 <= nj < M:
                        total += img[ni][nj]
                        count += 1

            result[i][j] = total // count  # làm tròn xuống

    # In ảnh sau khi làm mịn
    for row in result:
        print(*row)
```
