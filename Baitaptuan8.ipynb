```python
'''W7A16
Cho một mảng số nguyên không âm. Gọi
mảng đó là "mảng đẹp" nếu mỗi phần tử
x trong mảng xuất hiện đúng x lần.
Kiểm tra xem mảng có phải là mảng đẹp
hay không. Nếu không hãy xóa đi y số (y
là nhỏ nhất có thể) để mảng còn lại thành
mảng đẹp.'''
A = list(map(int, input().split()))
mydict = {}
for x in A:
    if x in mydict:
        mydict[x] += 1
    else:
        mydict[x] = 1
for keys in mydict.keys():
    if keys == 0:
        for i in range(mydict[keys]):
            A.remove(0)
    elif keys != mydict[keys]:
        if mydict[keys] > keys:
            diff = mydict[keys] - keys
            for i in range(diff):
                A.remove(keys)
        if mydict[keys] < keys:
            for i in range(mydict[keys]):
                A.remove(keys)
print(A)

        
```

```python
#W7A18
dayso = list(map(int, input().split()))
daychan = []
dayle = []
daycuoi =[]
for i in range(len(dayso)):
    if dayso[i] % 2 == 0:
        daychan.append(dayso[i])
    else:
        dayle.append(dayso[i])
for so in sorted(daychan):
    daycuoi.append(so)
for so in sorted(dayle, reverse=True):
    daycuoi.append(so)
for so in daycuoi:
    print(so, end=' ')
```
