```python
#Cau9
def nhapdict(tendict):
    n = input()
    pairs = n.split()
    tendict ={}
    for pair in pairs:
        key, value = pair.split(":")
        tendict[key] = int(value)
    return tendict
dict1 = nhapdict("dict1")
dict2 = nhapdict("dict2")
for key, value in dict2.items():
    if key in dict1:
        dict1[key] += value
    else:
        dict1[key] = value
print(dict1)




```

```python
#Cau10
ds = list(map(int, input().split()))
k = int(input("Nhập số k: "))
ketqua = []
loai = set()
for so1 in ds:
    ds1 = ds.copy()
    ds1.remove(so1)
    for so2 in ds1:
        if so1 + so2 == k:
            sonho = min(so1, so2)
            solon = max(so1, so2)
            cap = (sonho, solon)
            if cap not in loai:
                loai.add(cap)
                ketqua.append(cap)
ketqua.sort()
print(ketqua)
print(loai)

```

```python
#Cau12
ds = list(map(int, input().split()))
A = {}
for so in ds:
    if so in A:
        A[so] += 1
    else:
        A[so] = 1
print(A)
max = 0
B= []
for i in A.values():
    if i > max:
        max = i
for key, value in A.items():
    if value == max:
        B.append(key)
B.sort()
print("Các số xuất hiện nhiều nhất:", B[0])



    
    
            
```

```python
#Cau15
n = input().split()
k = int(input())
A = {}
B = {}
for item in n:
    key, value = item.split(':')
    A[key] = value
print(A)
for key,value in A.items():
    if int(value) > k:
        B[key] = int(value)
print(B)
```
