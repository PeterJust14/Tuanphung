```python
n=100 
def sokhacnhau(n):
    a=str(n)
    b=len(a)
    for i in range(b):
        for j in range(i+1,b):
            if a[i]==a[j]:
                return False
    return True
for i in range(1,int(n**0.5)+1):
    if i**2<=n and sokhacnhau(i):
        print(i**2, end =' ')
```

```python
def fib(n):
    A = [1,1]
    for i in range(2,n):
        A.append(A[i-1]+A[i-2])
    return A
print(fib(10))
```
