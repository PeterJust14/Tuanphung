```python
#Cau1
def Max_of_Two(a,b):
    max = 0
    if a>b:
        max=a
    if a<b:
        max =b
    return max
a,b = map(int,input().split())
print(Max_of_Two(a,b))
```

```python
#Cau2
def swap(a,b):
    a,b = b,a
    return a,b
x,y = map(int,input().split())
print(swap(x,y))
```

```python
#Cau3
def snt(a):
    souoc =0
    for i in range(1,a+1):
        if a% i==0:
            souoc +=1
    if souoc == 2:
        return True
    else:
        return False
m = int(input())
if snt(m):
    print("là snt")
else:
    print("không là snt")

```

```python
#Cau4
def sohoanhao(a):
    tonguoc = 0
    for i in range(1,a):
        if a % i ==0:
            tonguoc += i
    if tonguoc == a:
        return True
m = int(input())
if sohoanhao(m):
    print("Là số hoàn hảo")
else:
    print("không laf số hoàn hảo")
```

```python
#Cau6
def giaithua(a):
    giai_thua =1
    for i in range(1,a+1):
        giai_thua *= i
    return giai_thua
m = int(input())
print(giaithua(m))
```

```python
#Cau7
def pheptoan(a,b,c):
    if c == "+":
        return a+b
    if c == "-":
        return a-b
    if c == "*":
        return a*b
    if c == "/":
        return a/b
a,b = map(int,input("Nhap 2 so").split())
c = input("Nhap phep toan:")
print(pheptoan(a,b,c))

```

```python
#Cau8
def hamming(a,b):
    return(bin(a^b).count("1"))
a,b = map(int,input().split())
print(hamming(a,b))
```

```python
#Cau9
def tongcs(a):
    sum = 0
    for c in str(a):
        sum += int(c)
    return sum
a = int(input())
print(tongcs(a))
```

```python
#Cau10
def dangcau(a,b):
    if len(a) != len(b):
        return False
    A ={}
    B ={}
    for x,y in zip(a,b):
        if x in A and A[x] != y:
            return False
        if y in B and B[y] != x:
            return False
        A[x] = y
        B[y] = x
    return True
so1,so2 = input().split()
print(dangcau(so1,so2))
```
